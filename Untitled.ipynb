```python
sales = pd.read_csv('company_sales_data.csv')
sales
```

```python
profit = pd.DataFrame({"month_number": sales["month_number"], "pfacecream" : sales["facecream"]*10, "pfacewash" : sales["facewash"]*10, 
                       "ptoothpaste": sales["toothpaste"]*10, "pbathingsoap": sales["bathingsoap"]*10, 
                       "pshampoo":sales["shampoo"]*10, "pmoisturizer": sales["moisturizer"]*10})

profit_total= pd.DataFrame({"month_number": sales["month_number"], "pfacecream" : sales["facecream"]*10, "pfacewash" : sales["facewash"]*10, 
                       "ptoothpaste": sales["toothpaste"]*10, "pbathingsoap": sales["bathingsoap"]*10, 
                       "pshampoo":sales["shampoo"]*10, "pmoisturizer": sales["moisturizer"]*10})

profit_total.loc["Total"] = profit.sum()

profit_total
```

```python
fig, axes = plt.subplots(figsize=(15,15))
x1 = profit["month_number"]
y1 = profit["pfacecream"]
plt.plot(x1, y1, label = "facecream")


x2 = profit["month_number"]
y2 = profit["pfacewash"]
plt.plot(x2, y2, label = "facewash")


x3 = profit["month_number"]
y3 = profit["ptoothpaste"]
plt.plot(x3, y3, label = "ptoothpaste")


x4 = profit["month_number"]
y4 = profit["pbathingsoap"]
plt.plot(x4, y4, label = "pbathingsoap")


x5 = profit["month_number"]
y5 = profit["pshampoo"]
plt.plot(x5, y5, label = "pshampoo")


x6 = profit["month_number"]
y6 = profit["pmoisturizer"]
plt.plot(x6, y6, label = "pmoisturizer")


plt.xlabel('numéro mois')
plt.ylabel("profit")
plt.legend()
#plt.figure(figsize=(15, 15))
#plt.figure(figsize=(cm_to_inch(15),cm_to_inch(10)))
#fig = plt.figure()
#fig.set_figheight(5)
plt.title("Profit par produit et par mois")
plt.grid()
plt.show()
```
